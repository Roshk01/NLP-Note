```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split 
```

```python
data = pd.read_csv("spam.csv" , encoding = "latin1")
data.head()
```

```python
data.drop(columns = ['Unnamed: 2' , 'Unnamed: 3', 'Unnamed: 4'], inplace = True)
data.head()
```

```python
data.rename( {'v2':'msg'}, axis= 'columns', inplace = True)
```

```python
data['verify'] = data['v1'].apply(lambda x:1 if x=='spam' else 0)
```

```python
data.drop(columns = ['v1'], inplace = True)
```

```python
data.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.msg, data.verify, test_size = 0.2)
X_test.shape
```

```python
v = CountVectorizer()
X_train_cv = v.fit_transform(X_train,y_train)
```

```python
X_train_cv.shape
```

```python
dir(v)
```

```python
v.vocabulary_
```

```python
v.get_feature_names_out()[6052]
```

```python
X_train_np = X_train_cv.toarray()
```

```python
X_train_np[0]
```

```python
np.where(X_train_np[0] != 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train_cv, y_train)
```

# Convert test section into count vector

```python
X_test_cv = v.transform(X_test)
```

```python
y_pred = model.predict(X_test_cv)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
emails = {
    "hey, can you come sooner.",
    " congratulation! you win 15 lakh price and a iphone."
}
```

```python
email_count = v.transform(emails)
model.predict(email_count)
```

# Using sklearn Pipeline model 

```python
from sklearn.pipeline import Pipeline
pipe_model = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
pipe_model.fit(X_train,y_train)
```

```python
pipe_y_pred = pipe_model.predict(X_test)
print(classification_report(y_test, pipe_y_pred))
```
